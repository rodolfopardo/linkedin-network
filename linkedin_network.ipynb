```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt 
from pylab import rcParams
rcParams['figure.figsize'] = 10, 7
```

```python
data = pd.read_csv('Connections.csv')
```

```python
#Completando registros

def limpia(x):
    data[x].fillna('Sin registro', inplace = True)
    return data

limpia('Company')
limpia('First Name')
limpia('Last Name')
limpia('Position')
```

```python
data['First Name'].value_counts().head().plot.bar();
plt.title('Nombres más comunes de mi Red', fontsize = 20)
plt.xlabel('Nombres')
plt.ylabel('Frecuencia')
plt.show()
```

```python
data['Last Name'].value_counts().head().plot.bar();
plt.title('Nombres más comunes de mi Red', fontsize = 20)
plt.xlabel('Nombres')
plt.ylabel('Frecuencia')
plt.show()
```

```python
fig = px.treemap(data, path=['Mi red', 'Company', 'Position'], width=1000, height=1000)
fig.show()
```
